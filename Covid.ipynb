```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
country_list = ['India', 'Korea, South', 'Germany', 'Spain', 'Italy', 'Brazil', 'Thailand', 'Singapore']
```

```python
cvd19_confirmed = pd.read_csv('time_series_covid19_confirmed_global.csv')
```

```python
cvd19_deaths = pd.read_csv('time_series_covid19_deaths_global.csv')
```

```python
cvd19_confirmed.set_index('Country/Region',inplace=True)
```

```python
cvd19_confirmed = cvd19_confirmed.drop(['Province/State'], axis=1)
cvd19_confirmed = cvd19_confirmed.drop(['Lat'], axis=1)
cvd19_confirmed = cvd19_confirmed.drop(['Long'], axis=1)
```

```python
cvd19_confirmed.head()
```

```python
confirmed_dataset = pd.DataFrame()
```

```python
for country in country_list:
    confirmed_dataset = pd.concat([confirmed_dataset, cvd19_confirmed[cvd19_confirmed.index==country]])
```

```python
transpose_confirmed = confirmed_dataset.transpose()
```
